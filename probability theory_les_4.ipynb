#### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].Найдите ее среднее значение и дисперсию.

```python
mu = (200 + 800)/2
sigma2 = (800 -200)** 2/12
print(f"M(x) среднее значение : {round(mu,3)} \n", f"D(X) дисперсия равна: {round(sigma2,3)} \n")
```

#### 2.О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

```python
import numpy as np
a = 0.5
sigma2 = 0.2
# выражаю правую границу из формулы дисперсии
b = np.sqrt(12*sigma2) + a
mu = (a+b)/2
print(f"b правая граница: {round(b,3)} \n",f"M(x) среднее значение : {round(mu,3)} \n" )
```

#### 3.Непрерывная случайная величина X распределена нормально и задана плотностью распределения
f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-(x+2)**2) / 32).
Найдите:
$$а). M(X)$$
$$б). D(X)$$
$$в). std(X) (среднее квадратичное отклонение)$$


Формула плотности распределения вероятности описывается формулой: $$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}}$$ где
$$a - M(X); {\sigma^2} - D(x); {\sigma} - std(X)$$
имеем $$f(x) = \frac{1}{4 \sqrt{2 \pi}} e ^ {- \frac{(x + 2)^2}{32}}$$
тогда
$$M(X) = -a = -2 $$
$$D(X) = 16 $$
$$std(X) = 4$$

#### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?
Т.к. $D(X)=25$, то $\sigma=\sqrt{25}=5$. Тогда отклонение будет равно: 
$$\frac{190-178}{5}=2.4\sigma$$
